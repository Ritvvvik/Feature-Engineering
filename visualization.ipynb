MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
#scatter
plt.scatter(x,y,c='b',marker='o')
plt.scatter(a,b,c='r',marker='o')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
y = x*x
```

```python
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*-')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
x = np.arange(0,11)
y = 3 * x + 5
plt.plot(x,y)
```

```python
np.pi
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.subplot(2,1,1)
x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x,y_sin)
plt.title('Sine Curve')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine Curve')
```

```python
x = [2,6,10]
y = [12,14,16]
plt.bar(x,y)
plt.title('Simple Bar Graph')
```

```python
x = np.random.randint(1,100,10)
plt.hist(x)
```

```python
x = [np.random.normal(0,std,100) for std in range(1,4)]
x
```

```python
plt.boxplot(x,patch_artist=True)
```

```python
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow = False)
```
